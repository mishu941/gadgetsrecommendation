```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('C_LAPTOPS.csv')
```

```python
dataset.head(1)
```

```python
# Count the number of laptops for each brand
brand_count = dataset['Brand'].value_counts()

# Create a bar chart of the brand count
plt.bar(brand_count.index, brand_count.values)

# Set the chart title and axis labels
plt.title('Number of Laptops by Brand')
plt.xlabel('Brand')
plt.ylabel('Number of Laptops')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the chart
plt.show()
```

```python
#countplot brand column
sns.set_style('darkgrid')
sns.countplot(x='Brand', data=dataset)
plt.xticks(rotation=90)
plt.show()
```

```python
#create a heatmap of the correlation matrix for all numerical columns in the dataset, showing the pairwise correlations between different features.
corr_matrix = dataset.corr()
sns.set_style('darkgrid')
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
